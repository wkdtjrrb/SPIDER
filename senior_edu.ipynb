2. 정보화 기기를 이용할 수 있는 능력이 있는가?
- PC 이용 능력_1) 나는 필요한 프로그램(소프트웨어)을 컴퓨터에 설치/삭제/업데이트 할 수 있다
- PC 이용 능력_2) 나는 PC에 유선 또는 무선 인터넷을 스스로 연결해서 사용할 수 있다
- 문4. PC 이용 능력_3) 나는 웹 브라우저(크롬, 사파리, 엣지, 웨일 등)에서 내가 원하는 환경을 설정할 수 있다(팝업창 차단, 텍스트 크기 설정, 보안 및 시작 홈페이지 설정 등)
- PC 이용 능력_4) 나는 PC에 다양한 외장기기(디지털 카메라, 프린터, 스캐너, USB 외장하드 등)를 연결하여 이용할 수 있다
- PC 이용 능력_5) 나는 PC에 있는 파일을 인터넷을 통해 다른 사람에게 전송할 수 있다
- 문4. PC 이용 능력_6) 나는 PC의 악성코드(바이러스, 스파이웨어 등)를 검사/치료할 수 있다
- 문4. PC 이용 능력_7) 나는 PC에서 문서(한글, 엑셀, 파워포인트 등) 등을 작성할 수 있다
- 문5. 모바일기기 이용 능력_1) 나는 스마트기기에서 디스플레이/소리/보안/알림/입력방법 등의 환경설정을 할 수 있다
- 문5. 모바일기기 이용 능력_2) 나는 스마트기기에서 무선 랜(와이파이, 기가와이파이 포함) 설정을 할 수 있다
- 문5. 모바일기기 이용 능력_3) 나는 스마트기기에 있는 파일을 PC로 옮길 수 있다
- 문5. 모바일기기 이용 능력_4) 나는 스마트기기에 있는 파일/사진 등을 다른 사람에게 전송할 수 있다
- 문5. 모바일기기 이용 능력_5) 나는 앱을 스마트기기에 설치/삭제/업데이트할 수 있다
- 문5. 모바일기기 이용 능력_6) 나는 스마트기기의 악성코드(바이러스, 스파이웨어 등)를 검사/치료할 수 있다
- 문5. 모바일기기 이용 능력_7) 나는 스마트기기에서 문서(메모장, 워드)를 작성할 수 있다

3. 최근 정보화 기기를 얼마나 이용하였는가?
- 문7. 최근 한 달 이용 일수_1) PC(데스크탑, 노트북)
- 문7. 최근 한 달 이용 일수_2) 스마트폰
- 문7. 최근 한 달 이용 일수_3) 스마트패드(태블릿PC)

-> 정보화 기기를 사용하는 노인 :
스마트기기 중 하나라도 가지고 있는 노인이면서
정보화기기 능력의 평균이 2.5를 넘으면서
최근 한달 이내에 인터넷을 이용한 사람


```python
df.shape
```

```python
# 스마트 기기를 가진 사람 (피처폰 제외)
df1 = df[(df['Q3'] == 1) & ((df['Q1A1'] == 1) | (df['Q1A2'] == 1) | (df['Q2A11'] == 1) | (df['Q2A12'] == 1)
| (df['Q2A13'] == 1) | (df['Q2A2'] == 1) | (df['Q2A3'] == 1))]
df1.shape
```

```python
# 정보화 기기 능력의 평균이 2.5보다 큰 사람
df2 = df1[(df1['Q4A1'] + df1['Q4A2']+ df1['Q4A3']+ df1['Q4A4']+ df1['Q4A5']+
           df1['Q4A6']+ df1['Q4A7']+ df1['Q5A1']+ df1['Q5A2']+ df1['Q5A3']+
           df1['Q5A4']+ df1['Q5A5']+ df1['Q5A6']+ df1['Q5A7'])/13 >=2.5]
df2.shape
```

```python
# 평균 PC 이용 일수
df2['Q7A1'].mean()
```

```python
# 평균 스마트폰 이용 일수
df2['Q7A2'].mean()
```

```python
# 평균 스마트패드 이용 일수
df2['Q7A3'].mean()
```

```python
# 평균 스마트 기기 이용 일수가 평균 이상인 사람 / 정보화 노인
df_edu = df2[(df2['Q7A1']>8)|(df2['Q7A2']>28)|(df2['Q7A3']>1)]
df_edu.shape
```

```python
# 위의 세가지 필터에 맞지 않는 사람(정보화되지 않은 노인)
df_non_edu = df[~df.isin(df_edu.to_dict('list')).all(1)]
df_non_edu.shape
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
```

```python
# 정보화 노인의 삶의 만족도(1~4)
col1 = ['Q19A1','Q19A2','Q19A3','Q19A4','Q19A5','Q19A6','Q19A7','Q19A8','Q20A1','Q20A2','Q20A3','Q20A4','Q20A5']
df5 = df_edu[col1]

# 각 열의 평균 계산
column_means = df5.mean()

# 가로축 레이블 설정
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M']
column_means.index = labels

plt.figure(figsize=(10, 6))
ax = column_means.plot(kind='bar', color='skyblue')
plt.title("정보화 노인의 삶의 만족도")
plt.xlabel("항목")
plt.ylabel("평균")
plt.xticks(rotation=45)

# 전체 평균 값 빨간 점선으로 표시
overall_mean = column_means.mean()
ax.axhline(y=overall_mean, color='red', linestyle='--', label=f'전체 평균')

plt.legend()
plt.show()

# 각 칼럼의 전체 평균 출력
print("각 칼럼의 전체 평균:")
print(column_means)

# 전체 평균 출력
print("전체 평균:")
print(overall_mean)
```

```python
# 비정보화 노인의 삶의 만족도
df6 = df_non_edu[col1]

# 각 열의 평균 계산
column_means = df6.mean()

# 가로축 레이블 설정
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M']
column_means.index = labels

plt.figure(figsize=(10, 6))
ax = column_means.plot(kind='bar', color='skyblue')
plt.title("비정보화 노인의 삶의 만족도")
plt.xlabel("항목")
plt.ylabel("평균")
plt.xticks(rotation=45)

# 전체 평균 값 빨간 점선으로 표시
overall_mean = column_means.mean()
ax.axhline(y=overall_mean, color='red', linestyle='--', label=f'전체 평균')

plt.legend()
plt.show()

# 각 칼럼의 전체 평균 출력
print("각 칼럼의 전체 평균:")
print(column_means)

# 전체 평균 출력
print("전체 평균:")
print(overall_mean)
```

**2. 노인들이 정보화 기기를 사용하지 않는 이유 &  노인들이 불편함을 느끼는 부분 분석과정**

- 디지털 기술에 대한 태도 분석
1. 디지털 기술은 유용하다
2. 디지털 기술은 내 삶을 편리하게 한다
3. 디지털 기술은 나에게 좋은 것이다.
4. 디지털 기술을 더 많이 이용하고 싶다.

- 디지털 기기 이용 효능감 분석
1. 디지털 기기를 배우는데 자신이 있다
2. 디지털 기기를 활용하는데 자신이 있다.
3. 디지털 기기의 사용방법을 빠르게 알아낼 수 있다.
4. 디지털 기기를 더 많이 이용하고 싶다.

```python
# 정보화 노인의 디지털 기술에 대한 태도 분석
col2 = ['Q17A1','Q17A2','Q17A3','Q17A4']
df7 = df_edu[col2]
df7_means = df7.mean()

# 가로축 레이블 설정
labels = ['A', 'B', 'C', 'D']

plt.figure(figsize=(10, 6))
ax = df7_means.plot(kind='bar', color='skyblue')
plt.title("정보화 노인의 디지털 기술에 대한 태도")
plt.xlabel("항목")
plt.ylabel("평균")
plt.xticks(range(len(labels)), labels)  # A, B, C, D로 레이블 설정

# 평균값 빨간 점선으로 표시
overall_mean = df7_means.mean()
ax.axhline(y=overall_mean, color='red', linestyle='--', label=f'전체 평균')

plt.legend()
plt.show()

# 각 칼럼의 전체 평균 출력
print("각 칼럼의 전체 평균:")
print(df7_means)

# 전체 평균 출력
print("전체 평균:")
print(overall_mean)
```
